```python
class Chandrayaan3:
    def __init__(self, x, y, z, direction):
        self.position = {'x': x, 'y': y, 'z': z}
        self.direction = direction

    def forward(self):
        if self.direction == 'N':
            self.position['y'] += 1
        elif self.direction == 'S':
            self.position['y'] -= 1
        elif self.direction == 'E':
            self.position['x'] += 1
        elif self.direction == 'W':
            self.position['x'] -= 1
        elif self.direction == 'Up':
            self.position['z'] += 1
        elif self.direction == 'Down':
            self.position['z'] -= 1

    def backward(self):
        if self.direction == 'N':
            self.position['y'] -= 1
        elif self.direction == 'S':
            self.position['y'] += 1
        elif self.direction == 'E':
            self.position['x'] -= 1
        elif self.direction == 'W':
            self.position['x'] += 1
        elif self.direction == 'Up':
            self.position['z'] -= 1
        elif self.direction == 'Down':
            self.position['z'] += 1

    def left(self):
        if self.direction == 'N':
            self.direction = 'W'
        elif self.direction == 'S':
            self.direction = 'E'
        elif self.direction == 'E':
            self.direction = 'N'
        elif self.direction == 'W':
            self.direction = 'S'

    def right(self):
        if self.direction == 'N':
            self.direction = 'E'
        elif self.direction == 'S':
            self.direction = 'W'
        elif self.direction == 'E':
            self.direction = 'S'
        elif self.direction == 'W':
            self.direction = 'N'

    def up(self):
            self.direction = 'Up'

    def down(self):
            self.direction = 'Down'

    def execute_commands(self, commands):
        print("Initial Position:", self.position)
        print("Initial Direction:", self.direction)

        for cmd in commands:
            prev_position = self.position.copy()
            prev_direction = self.direction
            
            if cmd == 'f':
                self.forward()
            elif cmd == 'b':
                self.backward()
            elif cmd == 'l':
                self.left()
            elif cmd == 'r':
                self.right()
            elif cmd == 'u':
                self.up()
            elif cmd == 'd':
                self.down()
            
            print(f"Command: {cmd} | Change in Coordinates:", {k: self.position[k] - prev_position[k] for k in self.position})
            print(f"Command: {cmd} | Change in Direction:", prev_direction, "to", self.direction)

        return self.position, self.direction


def main():
    x = int(input("Enter initial x-coordinate: "))
    y = int(input("Enter initial y-coordinate: "))
    z = int(input("Enter initial z-coordinate: "))
    direction = input("Enter initial direction (N, S, E, W, Up, Down): ")
    commands = input("Enter commands ('f', 'r', 'u', 'b', 'l' without spaces): ")

    craft = Chandrayaan3(x, y, z, direction)
    final_position, final_direction = craft.execute_commands(commands)

    print("Final Position:", final_position)
    print("Final Direction:", final_direction)


if __name__ == "__main__":
    main()
```
